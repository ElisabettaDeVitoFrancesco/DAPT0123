```python
# Funzione per creare una tabella incrociata (moltiplicando tutti gli elementi con tutti)
def tab_incrociata(sequenza):
    
    sequenza_tabella = []

    for n in sequenza:
        sequenza_molt = []

        for m in sequenza:
            sequenza_molt.append(n*m)

        #print(sequenza_molt)
        sequenza_tabella.append(sequenza_molt)

    return sequenza_tabella
```

```python
tab_incrociata([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
tab_incrociata([3, 3, 9])
```

# Esercizio 4


```python
def min_lista(lista):
    minimo = lista[0]
    
    for n in lista:
        if minimo > n:
            minimo = n
    return minimo
```

```python
min_lista([4, 5, 6, 3, 1, 9, -2])
```

```python
min_lista([4, 5, 6, 3, 1, 9])
```
