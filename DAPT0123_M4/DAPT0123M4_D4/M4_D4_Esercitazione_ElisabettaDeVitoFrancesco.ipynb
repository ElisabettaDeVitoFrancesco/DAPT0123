## Esercizio 1


```python
# Creare le liste con i valori delle spese e relativi nomi dei clienti
spesa_cliente = []
nome_cliente = []

conta = 0
limite = 6

while conta < limite:
    conta += 1
    spesa = float(input("Spesa del cliente : "))
    nome = str(input("Nome del cliente: "))

    spesa_cliente.append(spesa)
    nome_cliente.append(nome)
```

```python
# Valore massimo tra le spese
spesa_max = max(spesa_cliente)
spesa_max_index = spesa_cliente.index(max(spesa_cliente))
cliente_spesa_max = nome_cliente[spesa_max_index]

# Valore minimo tra le spese
spesa_min = min(spesa_cliente)
spesa_min_index = spesa_cliente.index(min(spesa_cliente))
cliente_spesa_min = nome_cliente[spesa_min_index]

# Valore medio tra le spese
spesa_mean = sum(spesa_cliente) / len(spesa_cliente)

# Valore mediano tra le spese
spesa_cliente_sorted = sorted(spesa_cliente, key = float)
l = len(spesa_cliente_sorted)
if l % 2 == 0:
    spesa_median = (spesa_cliente_sorted[l // 2 - 1] + spesa_cliente_sorted[l // 2]) / 2
else:
    spesa_median = spesa_cliente_sorted[l // 2]

print(f"\nLa spesa maggiore  di {spesa_max}  del cliente {cliente_spesa_max}")
print(f"\nLa spesa minore  di {spesa_min}  del cliente {cliente_spesa_min}")
print(f"\nLa spesa media  di {spesa_mean} ")
print(f"\nLa mediana della spesa  di {spesa_median} ")
```

## Esercizio 2



```python
# Creare una lista con while loop e input
lista = []

conta = 0
limite = 6

while conta < limite:
    el = input(f"Elemento {conta} della lista: ")
    lista.append(el)
    conta += 1

# Creare una nuova lista, estraendo solo gli elementi degli indici dispari della lista precedente
lista_dispari = []
for n in range(len(lista)):
    #print(n)
    if n % 2 == 0:
        lista_dispari.append(lista[n])

print(f"\nLa nuova lista con solo gli elementi degli indici dispari  {lista_dispari}")
```

### Esercizio 3


```python
# Creare due sequenze con while loop e input
conta = 0
limite = 4

seq_1 = []
seq_2 = []

while conta < limite:
    conta += 1
    elem_seq_1 = int(input(f"Elemento {conta} della sequenza 1: "))
    elem_seq_2 = int(input(f"Elemento {conta} della sequenza 2: "))
    
    seq_1.append(elem_seq_1)
    seq_2.append(elem_seq_2)

# Stampare a video le due sequenze
print(f"\nLa sequenza 1  {seq_1}")
print(f"La sequenza 2  {seq_2}")

# Moltiplicare ogni elemento della seq_1 per ogni secondo elemento della seconda sequenza (== indici dispari della sequenza)
seq_molt = []
for s1 in seq_1:
    for i2 in range(len(seq_2)):
        if i2 % 2 != 0:
            prod = s1 * seq_2[i2]
            seq_molt.append(prod)

print(f"\nLa nuova sequenza degli elementi delle sequenze moltiplicate tra loro  {seq_molt}")

seq_molt_sum = sum(seq_molt)

print(f"\nLa somma di tutti gli elementi delle sequenze moltiplicate tra loro  {seq_molt_sum}")
```

## Esercizio 4


```python
# Selezionare un numero come input e creare una sequenza di numeri interi da 1 al numero di input, salvandola in una lista
conta = 0
seq_int = []

N = int(input("Selezionare un numero intero: "))

while conta < N:
    conta += 1
    seq_int.append(conta)

print(f"\nLa sequenza dei numeri interi da 1 a {N}  {seq_int}")

# Controllare quale dei numeri della lista di numeri interi sono fattori, con for loop e modulo
fattori = []
for n in seq_int:
    if N % n == 0:
        fattori.append(n)
print(f"I fattori di {N} sono {fattori}")
```
